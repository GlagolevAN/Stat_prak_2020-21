## Примеры использования функции distplot библиотеки Seaborn.

```python
import seaborn as sns 
from scipy.stats import norm 
import matplotlib.pyplot as plt 
```

```python
# На примере dataset-а tips.
df_tips = sns.load_dataset('tips')
df_tips.head(10)
```

```python
# Вызов функции distplot без дополнительных параметров.
sns.distplot(df_tips['tip']);
```

Построена гистограмма и ЯОП - ядерная оценка плостности (KDE - kernel density estimation).

### Наиболее важные параметры функции sns.distplot()

* a -  набор данных;
* bins - количество "корзин";
* hist - можно отключить отображение гистограммы (по умолчанию True);
* kde - отключить отображение ядерной оценки плотности (по умолчанию True);
* fit - подгонка распределения под имеющиеся данные;


```python
# Импортируем нормальное распределение, отключим отображение ЯОП.
from scipy.stats import norm
sns.distplot(df_tips['tip'], fit = norm, kde = False);
```

```python
# Импортируем гамма распределение, отключим отображение ЯОП.
from scipy.stats import gamma
sns.distplot(df_tips['tip'], fit = gamma, kde = False);
```

```python
# Можно сравнить приближение и ЯОП, иногда это может быть полезно.
sns.distplot(df_tips['tip'], fit = norm);
```

```python
#Найстройка цвета графика и разворот ориентации.
sns.distplot(df_tips['tip'], color = 'b', vertical=True);
```

```python
# Задание размеров
plt.figure(figsize = (8,6))
# Подпись оси 0x и подзаголовок графика.
sns.distplot(df_tips['tip'], axlabel = 'tip', label = 'Tips') 
# Задание заголовка графика
plt.title('Histogram of tips')
# Отображение подзаголовка (label) 
plt.legend()
plt.show()
```

Можно настраивать и другие параметры.

На графике ниже гистограмма и график подогнанного распределения имеют разные:

* цвета (у гистограммы есть еще параметр 'edgecolor', отсутствующий по понятным причинам у графика плотности распределения);
* тип линий (параметр 'linestyle');
* прозрачность (параметр 'alpha');
* толщину линий (параметр 'linewidth').

```python
# hist_kws - настройка параметров отображения гистограммы
# fit_kws - настройка параметров отображения подогнанного распределения
plt.figure(figsize = (8,6)) # задание размеров
sns.distplot(df_tips['tip'], fit = gamma, kde = False,
             hist_kws = {'color':'b', 'edgecolor':'r',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.5}, 
             fit_kws = {'color':'y', 'linewidth':4, 
                        'linestyle':'-', 'alpha':0.8});
```
