```python
m0 = 100

for k in [1, 2, 10]:
    for n in [10, 50, 200]:
        ps = []; 
        
        for j in range(m0):
            xs = []
            for i in range(n): 
                xs.append((stat.uniform().rvs(k).sum() - k * 0.5) *  2 * m.sqrt(3/k)) #1 2
            ps.append(stat.kstest(xs, stat.norm().cdf, N = n).pvalue) #2
        #print(sorted(ps))     
        plt.plot(sorted(ps), np.linspace(0, 1, m0))
        plt.plot(np.linspace(0, 1, m0), np.linspace(0, 1, m0))
        #sb.distplot(xs)
        plt.title('n = ' + str(n) + ' k = '+ str(k))
        plt.show()
```
