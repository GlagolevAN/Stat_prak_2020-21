```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb

#plt.style.use('ggplot')

my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
def generate(n):
    sample = []
    for i in range(n):
        theta = stat.uniform.rvs()
        sample.append(stat.bernoulli(theta).rvs())
    return np.array(sample)

x = np.linspace(0, 1, 1000)

for n in [5, 10, 20, 50, 100]:
    
    sample = generate(n)
    sum = sample.sum()
    plt.title("Vyborki")
    plt.plot(x, stat.beta(sum + 1, n - sum + 1).pdf(x), label = 'n = ' + str(n))
    plt.legend()
#plt.plot(x, stat.cauchy(loc = 3).pdf(x))
```

```python
x = np.linspace(0.5, 1.2, 1000)
for n in [5, 10, 20, 50, 100]:
    plt.title("1) $sum = 0.9n$ 2) $sum = 0.99n$")
    sum = n * 0.9
    plt.plot(x, stat.beta(sum + 1, n - sum + 1).pdf(x), label = '1) n = ' + str(n))
    
    sum = n * 0.99
    plt.plot(x, stat.beta(sum + 1, n - sum + 1).pdf(x), label = '2) n = ' + str(n))

    plt.legend()
```
