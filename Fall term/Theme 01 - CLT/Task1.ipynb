```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat

expon = lambda n: stat.gamma(n, loc = 1)
my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
x = np.linspace(0.01, 20, 100)

for i in range(10):
    plt.plot(x, expon(i+1).pdf(x))
```

Все плотности на одном графике

```python
x2 = np.linspace(-3, 3, 1000)
plt.plot(x2, my_norm(0, 1).pdf(x2))
for i in range(20):
    plt.plot(x2, m.sqrt(i + 1) * expon(i+1).pdf(m.sqrt(i + 1) * x2 + i + 1))
```

Плотности на разных графиках

```python
x2 = np.linspace(-5, 5, 1000)

for i in range(20):
    plt.plot(x2, m.sqrt(i + 1) * expon(i+1).pdf(x2 * m.sqrt(i + 1) + i + 1))
    plt.plot(x2, my_norm(0, 1).pdf(x2))
    plt.show()
```

Функции распределения

${\bf E} (X- {\bf E}X)^3 = {\bf E}X^3 - 3{\bf E}X^2 + 3{\bf E}X - 1 = 6 - 3 * 2 + 3 - 1 = 2$

$({\bf E}X^3) = \big(\dfrac{1}{1 - t}\big)\prime\prime\prime \big |_{t = 0} = 6$

${\bf D} X = 1$

Я посчитал ${\bf E} (X - {\bf E} X)^3$, который по модулю меньше, чем
${\bf E} |X - {\bf E} X|^3$. Даже с таким параметром функция распределения суммы лежит в границах неравенства Берри-Эссеена

```python
x2 = np.linspace(-5, 5, 1000)
const_B_E = 0.4784
for i in range(20):
    plt.plot(x2, expon(i+1).cdf(x2 * m.sqrt(i + 1) + i + 1))
    
    delta_f = const_B_E * 2 / m.sqrt(i+1)
    plt.plot(x2, my_norm(0, 1).cdf(x2) + delta_f, color = 'green')
    plt.plot(x2, my_norm(0, 1).cdf(x2))
    plt.plot(x2, my_norm(0, 1).cdf(x2) - delta_f, color = 'green')
    
    plt.title(i+1)
    plt.show()
    
```

```python
i = 999
plt.plot(x2, expon(i+1).cdf(x2 * m.sqrt(i + 1) + i + 1))
    
plt.plot(x2, my_norm(0, 1).cdf(x2) + const_B_E * 2 / m.sqrt(i+1), color = 'green')
plt.plot(x2, my_norm(0, 1).cdf(x2))
plt.plot(x2, my_norm(0, 1).cdf(x2) - const_B_E * 2 / m.sqrt(i+1), color = 'green')
    
plt.title(i+1)
plt.show()
```
