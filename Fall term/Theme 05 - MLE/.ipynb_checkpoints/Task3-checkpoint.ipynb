```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import seaborn as sb
import scipy.optimize as opt
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb


expon = lambda n: stat.gamma(n, loc = 1)
my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
def likelihood(distr, sample):
    L = 0
    for i in range(len(sample)):
        L += np.log(distr(sample[i]))
    return L; #return log of the likelihood

def sampling_method(distr, sample):
    D = distr(sample[0])
    for i in range(1, len(sample) - 1):
        D *= (distr(sample[i]) - distr(sample[i - 1]))
    D *= (1 - distr(sample[len(sample) - 1]))
    return m.exp(D) * 10 **(len(sample));

def likelihood_beta(a, b):
    return -likelihood(distr = stat.beta(a, b).pdf, sample = mini_sample)

def sampling_method_beta(a, b):
    return -sampling_method(distr = stat.beta(a, b).cdf, sample = mini_sample)
```

```python
samplesize = 1000

for n, p in [(100, 0.04), (1000, 0.004), (10, 0.4)]:

    sample1 = stat.binom.rvs(n, p, size = samplesize)
    sample2 = stat.poisson.rvs(4, size = samplesize)

    print("n =", n)
    plt.figure(figsize = (14, 6))
    plt.subplot(121); sb.distplot(sample1); plt.title('Binom')
    plt.subplot(122); sb.distplot(sample2, color = 'orange'); plt.title('Poisson'); 
    plt.show()

    plt.figure(figsize = (8, 6))
    sb.distplot(sample1, label = 'binom'); sb.distplot(sample2, label = "Poisson", color = 'orange')
    plt.title("Both for n = " + str(n) + ', p = ' + str(p))
    plt.legend(fontsize = 'x-large');
    plt.show()
```
