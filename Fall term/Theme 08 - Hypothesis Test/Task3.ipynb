```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb

#plt.style.use('ggplot')

my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

$X_{(n)}$ имеет ф.р. $(x - \theta)^n I_{x \in (\theta, \theta + 1)} + I_{x > \theta + 1}$ 

$X_{(1)}$ имеет ф.р. $1 - (1 -(x - \theta))^n I_{x \in (\theta, \theta + 1)} + I_{x > \theta + 1}$

```python
def f_n(x):
    if (theta0 < x < theta0 + 1):
        return (x - theta0)**n
    if (x >= theta0 + 1):
        return 1
    return 0

def f_1(x):
    if (theta0 < x < theta0 + 1):
        return 1 - (1 - (x - theta0))**n
    if (x >= theta0 + 1):
        return 1
    return 0
```

# theta1 = 0.05

```python
theta0 = 0
#theta1_axis = np.linspace(-5, 5, 1000)
theta1 = 0.05
alpha = 0.05
m = 150
x_axis = np.linspace(0, 1, m)
for n in [10]:
    max_sample = []; min_sample = []; T1_sample = []; T2_sample = []; T3_sample = [];
    T1_sample1 = []; T2_sample1 = []; T3_sample1 = [];
    for ssize in range(m):
        sample = stat.uniform(theta0).rvs(size = n)
        min_sample.append(sample.min())
        max_sample.append(sample.max())
        
        T1_sample.append(1 - f_1(sample.min()))
        T2_sample.append(f_1(sample.min()))
        T3_sample.append(1 - f_n(sample.max()))
    
        sample1 = stat.uniform(theta1).rvs(size = n)
        #min_sample1.append(sample1.min())
        #max_sample1.append(sample1.max())
        
        T1_sample1.append(1 - f_1(sample1.min()))
        T2_sample1.append(f_1(sample1.min()))
        T3_sample1.append(1 - f_n(sample1.max()))
        #plt.plot(x_axis, (1 - stat.uniform(theta0, 1).cdf(x_axis))**n, label = '1 Род')
        #print(sample.max())
        #print(1- f_1(sample.min()))
        #print()
    #print(T1_sample)
    T1_sample = sorted(T1_sample); T1_sample1.sort()
    T2_sample.sort(); T2_sample1.sort()
    T3_sample.sort(); T3_sample1.sort()
    #print(T1_sample)
    
    #три критерия при нулевой гипотезе
    
    plt.plot(x_axis, T1_sample, label = '$T_1$ ')
    plt.plot(x_axis, T2_sample, label = '$T_2$ ')
    plt.plot(x_axis, T3_sample, label = '$T_3$ ')
    plt.legend()
    plt.title('При $\\theta = \\theta_0$')
    plt.show()
    
    #theta_1
    plt.figure(figsize = (9, 9))
    plt.plot(x_axis, T1_sample1, label = '$T_1$ при $\\theta_1$')
    plt.plot(x_axis, T2_sample1, label = '$T_2$ при $\\theta_1$')
    plt.plot(x_axis, T3_sample1, label = '$T_3$ при $\\theta_1$')
    plt.legend()
    plt.title('При $\\theta = \\theta_1$')
    plt.show()
    
    #all togheter
    plt.figure(figsize = (9, 9))
    plt.plot(x_axis, T1_sample, label = '$T_1$ при $\\theta_0$')
    plt.plot(x_axis, T2_sample, label = '$T_2$при $\\theta_0$')
    plt.plot(x_axis, T3_sample, label = '$T_3$ при $\\theta_0$')
    plt.plot(x_axis, T1_sample1, label = '$T_1$ при $\\theta_1$')
    plt.plot(x_axis, T2_sample1, label = '$T_2$ при $\\theta_1$')
    plt.plot(x_axis, T3_sample1, label = '$T_3$ при $\\theta_1$')
    plt.legend()
    plt.title('При $\\theta = \\theta_0$ и $\\theta_1$')
    plt.show()
```
