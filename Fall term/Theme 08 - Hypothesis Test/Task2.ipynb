```python
theta0 = 0
theta1_axis = np.linspace(-5, 5, 1000)
alpha = 0.05

for n in [5, 20, 50, 100]:
        
        #plt.plot(x_axis, (1 - stat.uniform(theta0, 1).cdf(x_axis))**n, label = '1 Род')
    C = stat.uniform(theta0, 1).ppf(1 - alpha**(1/n))
        
    plt.plot(theta1_axis, (1 - stat.uniform(theta1_axis, 1).cdf(C))**n, label = '$n = $' + str(n))
    plt.legend()
    plt.title('$X_{(1)}$ > C')
plt.show()
        
```

```python
for n in [5, 20, 50, 100]:
    C = stat.uniform(theta0, 1).ppf(1 - (1 - alpha)**(1/n))
    
    plt.plot(theta1_axis, 1 - (1 - stat.uniform(theta1_axis, 1).cdf(C))**n, label = '$n = $' + str(n))
    plt.legend()
    plt.title('$X_{(1)} < C$')
plt.show()
        
```
