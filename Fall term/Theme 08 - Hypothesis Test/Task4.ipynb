```python
theta0 = 0.5
x_axis = np.linspace(0, 1, 1000)



for theta1 in [1/3]:
    for n in [5, 10, 20, 50, 100]:
        
        plt.plot(x_axis, stat.binom(n, theta0).cdf(n * x_axis), label = '1 Род')
        plt.plot(x_axis, 1 - stat.binom(n, theta1).cdf(n * x_axis), label = '2 Род')
        plt.legend()
        plt.title('$\\overline{X}$ < C')
        plt.show()
```
