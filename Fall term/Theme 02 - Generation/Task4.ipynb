```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb


expon = lambda n: stat.gamma(n, loc = 1)
my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
def length(x, y):
    return abs(x) + y;

def vyborka_triangle(size):
    counter = 0; sample_x = []; sample_y = []
    while (counter < size):
        x = stat.uniform.rvs(-1, 2)
        y = stat.uniform.rvs()
        if (length(x, y) <= 1):
            counter += 1
            sample_x.append(x)
            sample_y.append(y)
    return sample_x, sample_y
```

```python
sample_x, sample_y = vyborka_triangle(1000)

plt.figure(figsize=(8, 8))

x_beauty = np.linspace(-1, 0, 1000)

y_beauty = [m.sqrt(1 - element**2) for element in x_beauty]
plt.plot(x_beauty, [1 + elem for elem in x_beauty], color = 'k', label = 'true triangle')
x_beauty = np.linspace(0, 1, 1000)
plt.plot(x_beauty, [1 - elem for elem in x_beauty], color = 'k')



plt.plot(sample_x, sample_y, 'o', color = 'green', label = 'in triangle')

plt.xscale("linear")
plt.yscale("linear")
plt.axis([-1, 1, 0, 1])
plt.legend(loc = 'upper left')
plt.title("Uniform in triangle")
plt.show()
```
