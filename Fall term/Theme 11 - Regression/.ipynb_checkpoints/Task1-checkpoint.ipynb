```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb
from sklearn.linear_model import LinearRegression
#plt.style.use('ggplot')

my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
a = 3; b = 3;
```

```python
a = 3; b = 3;
for m0 in [10, 20, 50]:
    xs = np.arange(1 / m0, 1, step = 1 / m0)
    ys = np.array([a * elem + b + my_norm(0, 1).rvs() for elem in xs])
    X = np.array([ [elem, 1] for elem in xs])
    result = LinearRegression().fit(X, ys)
    a_pred = result.coef_[0]
    b_pred = result.intercept_
    
    print('m =', m0)
    print('a =', a_pred)
    print('b =', b_pred)
    print('covarience matrix of the resultant normal distr: \n', np.linalg.inv(X.T@X))
    print('Its eigenvalues: \n', np.linalg.eigvals(np.linalg.inv(X.T@X)))
    print('sigma = RSS / m = ' + str(np.linalg.norm(ys - X@[a_pred, b_pred]) ** 2 / m0))
    
    
    plt.scatter(xs, ys)
    plt.plot(xs, [a * elem + b for elem in xs] , label = 'true params', color = 'green')
    plt.plot(xs, [a_pred * elem + b_pred for elem in xs] , label = 'estimated params', color = 'red')
    
    plt.legend()
    plt.title('m = ' + str(m0) + '  $\\sigma = 1$')
    plt.show()
```
