```python
keys = np.arange(200, 420 ,10)

sum_scores_2 = list(df[df['ГТО.оценка'] == 2] ['Сумма.баллов'])
discrete_2 = {elem: 0 for elem in keys}
for score in sum_scores_2:
    discrete_2[round(score, -1)] += 1
    
sum_scores_0 = list(df[df['ГТО.оценка'] == 0] ['Сумма.баллов'])
discrete_0 = {elem: 0 for elem in keys}
for score in sum_scores_0:
    discrete_0[round(score, -1)] += 1

gto_pvalue = stat.chi2_contingency([[discrete_0[elem] for elem in keys], [discrete_2[elem] for elem in keys]])[1]
```

```python
keys = np.arange(200, 420 ,10)

sum_scores_2 = list(df[df['Аттестат.с.отличием.оценка'] == 5] ['Сумма.баллов'])
discrete_2 = {elem: 0 for elem in keys}
for score in sum_scores_2:
    discrete_2[round(score, -1)] += 1
    
sum_scores_0 = list(df[df['Аттестат.с.отличием.оценка'] == 0] ['Сумма.баллов'])
discrete_0 = {elem: 0 for elem in keys}
for score in sum_scores_0:
    discrete_0[round(score, -1)] += 1

cert_pvalue = stat.chi2_contingency([[discrete_0[elem] for elem in keys], [discrete_2[elem] for elem in keys]])[1]
```

```python
keys = np.arange(200, 420 ,10)

sum_scores_gto_2 = list(df[df['Сочинение.оценка'] != 0] ['Сумма.баллов'])
discrete_gto_2 = {elem: 0 for elem in keys}
for score in sum_scores_gto_2:
    discrete_gto_2[round(score, -1)] += 1
    
sum_scores_gto_0 = list(df[df['Сочинение.оценка'] == 0] ['Сумма.баллов'])
discrete_gto_0 = {elem: 0 for elem in keys}
for score in sum_scores_gto_0:
    discrete_gto_0[round(score, -1)] += 1

essay_pvalue = stat.chi2_contingency([[discrete_gto_0[elem] for elem in keys], [discrete_gto_2[elem] for elem in keys]])[1]
```

```python
print('p-value for GTO badge', gto_pvalue)
print('p-value for certificate with blue ribbon', cert_pvalue)
print('p-value for essay', essay_pvalue)
```
