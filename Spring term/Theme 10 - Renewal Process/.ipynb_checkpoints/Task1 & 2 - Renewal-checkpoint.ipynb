```python
def task2(t, iters = 500):
    excesses = []; defects = []
    for it in range(iters):
        Arina_time = t

        bus_time = 0
        while (bus_time < Arina_time):
            deltat = stat.uniform.rvs(size = 1)[0]
            bus_time += deltat
        excesses.append(bus_time - Arina_time)
        #defects.append(Arina_time - bus_time + deltat)
    sb.distplot(excesses);
    print('Empirical excess time is', np.mean(excesses))
    #print('Theoretical time is', 1/3)
```
