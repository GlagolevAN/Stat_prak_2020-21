```python
def if_lived(k, n):
    
    flag = 0;
    while flag == 0:
        it = 0; p = 0.49;
        zs = [1]
        ys = [1]; z_k = 0;
        while (it < k) and (zs[-1] > 0):
            zs.append(sum(stat.geom(p).rvs(zs[-1])) - zs[-1])
            ys.append(ys[-1] + zs[-1])
            it += 1
        if zs[-1] > 0:
            flag = 1
    
    new_zs = 1; counter = 0
    for _ in range(zs[-1]):
        it = k
        while (it < n) and (new_zs > 0):
            new_zs = (sum(stat.geom(p).rvs(new_zs)) - new_zs)
            it += 1
        if it == n:
            counter += 1
    print('Answer is', counter/zs[-1])
```
