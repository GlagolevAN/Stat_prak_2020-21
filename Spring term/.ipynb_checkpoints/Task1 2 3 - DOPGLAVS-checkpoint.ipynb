### 1.1 a

Матожидание 0, дисперсия слагаемого 1, дисперсия суммы n.

$\displaystyle  \begin{array}{{>{\displaystyle}l}}
Y_{i} \ -\ независимые\ разнораспределенные\\
t_{n}^{2} =\sum _{i=1}^{n} DY_{i}\\
X_{n,i} =Y_{i} /t_{n}\\
S_{n} \ =\ \sum _{i=1}^{n} Y_{i} /t_{n}\\
Условие\ Линдеберга:\\
\dfrac{1}{t_{n}^{2}}\sum _{i=1}^{n} E\left( Y_{i}^{2} ;|Y_{i} | >\varepsilon t_{n}\right)\rightarrow 0,\ n\rightarrow \infty \ \forall \varepsilon  >0
\end{array}$

```python
def gen11a(n):
    summa = 0
    for i in range(1, n):
        a = list(stat.multinomial(1, [0.5/i/i, 1 - 1/i/i, 0.5/i/i]).rvs()[0]).index(1)
        if a == 0:
            summa -= i
        elif a == 2:
            summa += i
            
    return summa


def gen11b(n):
    summa = 0
    for i in range(1, n):
        a = list(stat.multinomial(1, [0.25, 0.5, 0.25]).rvs()[0]).index(1)
        if a == 0:
            summa -= i
        elif a == 2:
            summa += i
            
    return summa
```

$t^2_n = n$ 

берем $\varepsilon = 1$

$$
\dfrac{1}{t^2_n} \sum_{i=1}^n {\bf E} (Y^2_i; |Y_i| > \varepsilon t_n) = 
\dfrac{1}{n} \sum_{i=1}^n {\bf E} (Y^2_i; |Y_i| > \varepsilon \sqrt{n}) =
\dfrac{1}{n} \sum_{i=[\sqrt{n}]+1}^n (1/2) > \dfrac{1}{n} \sum_{i=n/2}^n (1/2) =  1/4 >0
$$

Условие Линдеберга не выполнено, значит нет сходимости к нормальному распределению

```python
n = 250
sample = []
for i in range(250):
    sample.append(gen11a(n)/m.sqrt(n))
        
sb.distplot(sample)
plt.show()
```

### 1.2 a

$\displaystyle F_{0}( x) =\ F_{X}( -x) +1-F_{X}( x) =L( x) /x^{\beta } ,x\rightarrow +\infty ,\ \beta \in ( 0,2]$

$\displaystyle b_{n} =F_{0}^{-1}( 1/n)$

В нашем случае:
$1 - 1/\sqrt{x} = y$

$x = \left(\dfrac{1}{1-y}\right)^2$ - для генерации

$F_0(x) = \dfrac{1}{2}(e^{-x} + x^{-1/2}) = 1/n \Rightarrow x \sim n^2$ (это $b_n$)

($\beta = 1/2$)

```python
def gen_sqrt(n):#распределение с ф.р. 1 - 1 / sqrt(x)
    precursor = stat.uniform().rvs(size = n);
    sample = [(1-y)**(-2) for y in precursor]
    return sample

def gen_12a(n):
    summa = 0
    rand = stat.binom(n, 0.5).rvs(size = 1)[0];
    summa -=  np.sum(stat.expon(1).rvs(size = rand))
    summa += np.sum(gen_sqrt(n - rand))
    
    return summa
    #for i in range(n):
    #    if (stat.bernoulli(0.5).rvs(size = 1)[0]):
    #        summa -= stat.expon(1).rvs(size = 1)[0]
    #    else:
    #        summa += stat.pareto().rvs()
```

# 1.2d

У нас симметричная схема

$\displaystyle F_(x) = 1 - L( x) /x^{\beta } ,x\rightarrow +\infty ,\ \beta = 3/2$

$\displaystyle b_{n} =F_{0}^{-1}( 1/n)$

В нашем случае:
$1 - 1/\sqrt{x^3} = y$

$x = \left(\dfrac{1}{1-y}\right)^{2/3}$ - для генерации

$F_0(x) = (x^{-3/2}) = 1/n \Rightarrow x \sim n^{2/3}$ (это $b_n$)

```python
def gen_32(n):
    precursor = stat.uniform().rvs(size = n);
    sample = [(1-y)**(-2/3) for y in precursor]
    return sample

def gen_12d(n):
    summa = 0
    rand = stat.binom(n, 0.5).rvs(size = 1)[0];
    summa -=  np.sum(gen_32(rand))
    summa += np.sum(gen_32(n - rand))
    
    return summa
```

# 1.3 a

$a = 3$

$\sigma^2 = 3$

   ${\bf E} |\xi_1 - a|^3 = \dfrac{6 (53 - e^3)}{e^3}$

Реальное отклонение (значение C)

```python
sups = []
nlist = [10, 50, 100, 1000]
for n in nlist:
    a = 3
    sigma = m.sqrt(3)
    sup = np.max([abs(stat.gamma(3 * n, scale = 1).cdf(x * sigma * m.sqrt(n) + n * a) - stat.norm().cdf(x))\
                    for x in np.linspace(-3 * m.sqrt(n), 5*m.sqrt(n), 60 * m.floor(m.sqrt(n)))])
    
    sups.append(sup * m.sqrt(n) *  sigma**3  * m.e**3 / (6 * (53 - m.e**3)))
        
plt.plot(nlist, sups)
plt.xlabel('n')
plt.ylabel('deviation')
plt.title('Gamma(3,1)')
plt.show()
print(sups)
```

##### Отклонение от ф.р. (абсолютное)

```python
real_errors = []
nlist = [10, 50, 100, 1000]
for n in nlist:
    a = 3
    sigma = m.sqrt(3)
    sup = np.max([abs(stat.gamma(3 * n, scale = 1).cdf(x * sigma * m.sqrt(n) + n * a) - stat.norm().cdf(x))\
                    for x in np.linspace(-3 * m.sqrt(n), 5*m.sqrt(n), 60 * m.floor(m.sqrt(n)))])
    
    real_errors.append(sup)
        
plt.plot(nlist, real_errors)
plt.xlabel('n')
plt.ylabel('deviation')
plt.title('Gamma(3,1)')
plt.show()
print('real errors:',real_errors)
plt.show()
```

```python
#деленые на 10 погрешности из неравенства Берри-Эссеена 
plt.plot(nlist, [0.4784 / m.sqrt(n) *0.1 /  sigma**3  / m.e**3 * (6 * (53 - m.e**3)) for n in nlist])
plt.plot(nlist, real_errors)
plt.xlabel('n')
plt.ylabel('deviation')
plt.title('Gamma(3,1)')
plt.show()
```

### Асимптотическое разложение

$\displaystyle  \begin{array}{{>{\displaystyle}l}}
F_{( S_{n} -\mu n) /\left( \sigma \sqrt{n}\right)}( x) -\Phi ( x) \ =\ \dfrac{\mu _{3}}{6\sigma ^{3}\sqrt{2\pi n}}\left( 1-x^{2}\right) \ e^{-x^{2} /2} +o\left(\dfrac{1}{\sqrt{n}}\right) ,\ где\ o() \ равномерно\ мало\\
по\ x\in \mathbb{R}
\end{array}$

```python
real_errors2 = []
nlist = [10, 50, 100, 1000]
mu_3 =  (6 * (53 - m.e**3))/m.e**3

for n in nlist:
    a = 3
    sigma = m.sqrt(3)
    sup = np.max([abs(stat.gamma(3 * n, scale = 1).cdf(x * sigma * m.sqrt(n) + n * a) - stat.norm().cdf(x)-\
                     mu_3 / (6 * sigma**3*m.sqrt(2*m.pi*n)) * (1-x**2) * m.exp(-x**2/2))\
                    for x in np.linspace(-3 * m.sqrt(n), 5*m.sqrt(n), 60 * m.floor(m.sqrt(n)))])
    
    real_errors2.append(sup)
    

plt.plot(nlist, real_errors, label = 'CLT')
plt.plot(nlist, real_errors2, label = 'asymp')
plt.xlabel('n')
plt.ylabel('deviation')
plt.title('Gamma(3,1)')
plt.legend()

plt.show()
```
