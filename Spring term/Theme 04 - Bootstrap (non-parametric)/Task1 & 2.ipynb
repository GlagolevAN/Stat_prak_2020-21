```python
mu = 5
```

```python
sample = my_norm(mu, 1).rvs(100)
```

```python
def get_mean(boot_sample, sample):
    summa = 0;
    for i in range(len(sample)):
        summa += boot_sample[i] * sample[i]
    return summa / len(sample)
```

```python
exp_sample = []

for i in range(1000):
    boot_sample = stat.multinomial(100, 0.01 * np.ones(100)).rvs()[0]
    exp_sample.append(m.exp(get_mean(boot_sample, sample)))
    
print('  Varience is '+str(np.var(exp_sample)))
print('  Interval is (' + str(np.mean(exp_sample) - 2 * m.sqrt(np.var(exp_sample))) + ', ' \
      + str(np.mean(exp_sample) + 2 * m.sqrt(np.var(exp_sample))) +')') #2 стандартных отклонения
print('True vaule is ' + str(m.exp(5)))
print('Estimated value ', np.mean(exp_sample))
```
