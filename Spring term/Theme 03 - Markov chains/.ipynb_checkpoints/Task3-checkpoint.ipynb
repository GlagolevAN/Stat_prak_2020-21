```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb
from sklearn.linear_model import LinearRegression
plt.style.use('ggplot')

my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
```

```python
A = np.array([[1/3, 1/3, 1/3], [0.2, 0.4, 0.4], [0.5, 1/6, 1/3]])
```

```python
for n in [1, 8, 30]:
    print('Xi_' + str(n) +' has distibution:')
    print([6/17, 5/17, 6/17]@np.linalg.matrix_power(A,n))
    #print(np.dot([6/17, 5/17, 6/17], np.linalg.matrix_power(A,n)))
    
```

```python
sample = []
for n in range(1):
    state = list(stat.multinomial(1, [6/17, 5/17, 6/17]).rvs()[0]).index(1) + 1
    for i in range(30):
        if (state == 1):
            state = list(stat.multinomial(1, A[0, :]).rvs()[0]).index(1) + 1
        elif (state == 2):
            state = list(stat.multinomial(1, A[1, :]).rvs()[0]).index(1) + 1
        elif (state == 3):
            state = list(stat.multinomial(1, A[2, :]).rvs()[0]).index(1) + 1
        sample.append(state)
    
plt.plot(np.arange(30), sample);
```
