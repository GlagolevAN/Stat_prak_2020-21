```python
def F_inv_x(x):
    '''inverse F_x'''
    if x > 0.9:
        return 1.5 - m.sqrt(10 - 10 * x)/2
    if x > 0.1:
        return (x - 0.5) / (0.4)
    #if x > 0
    return -1.5 + m.sqrt(10 * x)/2
```

```python
x_axes = np.linspace(0, 1, 1000)
plt.plot(x_axes, [F_inv_x(elem) for elem in x_axes])
```

```python
def F_x(x):
    ''' F_x'''
    if x > 1:
        return 1.2 * x - 0.4 * x * x + 0.1
    if x > -1:
        return 0.5 + 0.4 * x
    #if x > 0
    return 1.2 * x + 0.4 * x * x + 0.9
```

```python
x_axes = np.linspace(-1.5, 1.5, 1000)
plt.plot(x_axes, [F_x(elem) for elem in x_axes])
```
