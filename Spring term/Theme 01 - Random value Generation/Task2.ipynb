Хотим, чтобы $r * (r * \varphi) \sim const$ как малый элемент площади

Очевидно, что $\varphi \sim {\bf R}[0, 2 * pi]$

$r^2 \sim {\bf R}[1, 1.5]$

${\bf P} (r \leqslant x) = {\bf P} (r^2 \leqslant x^2) = \dfrac{x^2 - 1}{0.5} =: F(x)$

$y = \dfrac{x^2 - 1}{0.5}; \quad y/2 = x^2 - 1; \quad \sqrt{y/2 + 1} = x$

```python
def gen_n(n):
    gen_s = [[], []]
    for i in range(n):
        gen_s[0].append(stat.uniform().rvs() * 2 * m.pi)
        gen_s[1].append(m.sqrt(stat.uniform.rvs() * 0.5 + 1))
    return gen_s
```

```python
sample = gen_n(200)
#plt.polar(sample[1::2], sample[::2])

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='polar')
c = ax.scatter(sample[0], sample[1])
```
