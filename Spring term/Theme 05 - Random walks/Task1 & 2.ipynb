```python
def pmone():
    return 2 * stat.bernoulli(0.5).rvs() - 1
```

```python
print(pmone())
```

```python
sample = []
for i in range(100):
    summa = 0; beta = 0;
    for j in range(1000):
        summa += pmone()
        if (summa == 0):
            beta = j + 1;
    sample.append(beta)
```

```python
sample = np.array(sample)
sample.sort()
sample
```

```python
plt.plot(sample/1000, np.linspace(0, 1, 100))
xaxes = np.linspace(0, 1, 100)
plt.plot(xaxes, stat.arcsine.cdf(xaxes))
```
