# 2

Randomized

```python
excesses = []; defects = []
for it in range(iters):
    Arina_time = stat.uniform(loc = 20, scale = 50).rvs(1)[0]
    
    bus_time = 0
    while (bus_time < Arina_time):
        deltat = stat.expon(scale = 1 / lam).rvs(size = 1)[0]
        bus_time += deltat
    excesses.append(bus_time - Arina_time)
    defects.append(Arina_time - bus_time + deltat)
    
```

```python
print('Empirical waiting time is', np.mean(excesses))
print('Theoretical time is', 1/3)
```

```python
print('Empirical time from the last bus is', np.mean(defects))
print('Theoretical time is', 1/3)
```

```python
sb.distplot(excesses);
```
