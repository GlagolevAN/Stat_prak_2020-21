```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import math as m
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt
import seaborn as sb
from scipy.integrate import quad
from scipy.integrate import dblquad
from scipy.special import comb
from sklearn.linear_model import LinearRegression
plt.style.use('ggplot')

my_norm = lambda exp, disp: stat.norm(loc = exp, scale = m.sqrt(disp))
from mpmath import mp
#from mpmath import e
```

```python
n = 50; K = 5
```

### Нормальное

```python
fp = []; kwp = []
for i in range(100):
    sample = [stat.norm().rvs(n) for _ in range(K)]
    fp.append(stat.f_oneway(sample[0], sample[1], sample[2], sample[3], sample[4]).pvalue)
    kwp.append(stat.kruskal(sample[0], sample[1], sample[2], sample[3], sample[4]).pvalue)
    
fp.sort(); kwp.sort()
```

```python
plt.figure(figsize = (7,7))
plt.plot(fp, np.linspace(0, 1, 100), label = 'F-test')
plt.plot(kwp, np.linspace(0, 1, 100), label = 'Kruskal-Wallis')
plt.plot(np.linspace(0,1,100), np.linspace(0, 1, 100))
plt.legend()
plt.show()
```

### Распределение лапласа

```python
fp = []; kwp = []
for i in range(100):
    sample = [stat.laplace().rvs(n) for i in range(K)]
    #sample.append(stat.uniform(loc = -1, scale = 2).rvs(n))
    fp.append(stat.f_oneway(sample[0], sample[1], sample[2], sample[3], sample[4]).pvalue)
    kwp.append(stat.kruskal(sample[0], sample[1], sample[2], sample[3], sample[4]).pvalue)
    
fp.sort(); kwp.sort()

plt.figure(figsize = (7,7))
plt.plot(fp, np.linspace(0, 1, 100), label = 'F-test')
plt.plot(kwp, np.linspace(0, 1, 100), label = 'Kruskal-Wallis')
plt.plot(np.linspace(0,1,100), np.linspace(0, 1, 100))
plt.legend()
plt.show()
```

```python
counterf = 0; counterkw = 0 
for i in range(100):
    if fp[i] > 0.05:
        counterf += 1
    if kwp[i] > 0.05:
        counterkw += 1

print('f rejects', counterf, 'times')
print('kw rejects', counterkw, 'times')
```
